```python
import numpy as np
import random as rnd


matrizinicial = []
ancho= 5
alto = 2
#ancho= int(input('Ingrese el ancho de la matriz: '))
#alto = int(input('Ingrese el alto de la matriz: '))

for i in range(ancho * alto):
    matrizinicial.append(rnd.randint(1,5))

matriz1 = np.reshape(matrizinicial, (alto, ancho), order="C")

matrizFinal = [None for i in range(len(matrizinicial))]


def isTopLeftItem(i):
    if i == 0:
        return True
    else:
        return False


def isTopRigthItem(i):
    if i == ancho - 1:
        return True
    else:
        return False


def isBottomLeftItem(i):
    if i == len(matrizinicial) - ancho:
        return True
    else:
        return False


def isBottomRigthItem(i):
    if i == len(matrizinicial) - 1:
        return True
    else:
        return False


def isTopRow(i):
    if i > 0 and i < ancho - 1:
        return True
    else:
        return False


def isLeftColumn(i):
    if isTopLeftItem(i) is False and isBottomLeftItem(i) is False and i % ancho == 0:
        return True
    else:
        return False


def isRigthColumn(i):
    if isTopRigthItem(i) is False and isBottomRigthItem(i) is False and (i + 1) % ancho == 0:
        return True
    else:
        return False


def isBottomRow(i):
    if isBottomLeftItem(i) is False and isBottomRigthItem(i) is False and i >= len(matrizinicial) - ancho and i < len(matrizinicial):
        return True
    else:
        return False


def isCenterItem(i):
    if (i > 0 and i < len(matrizinicial) - 1 and isTopLeftItem(i) is False and isTopRow(i) is False and isTopRigthItem(i) is False and isLeftColumn(i) is False and isRigthColumn(i) is False and isBottomLeftItem(i) is False and isBottomRow(i) is False and isBottomLeftItem(i) is False):
        return True
    else:
        return False


for i in range(len(matrizinicial)):
    if isTopLeftItem(i):
        if matrizinicial[i] == matrizinicial[i + 1] or matrizinicial[i] ==matrizinicial[ancho]:
            matrizFinal[i] = 1
        else:
            matrizFinal[i] = 0
    if isTopRow(i):
        if matrizinicial[i] == matrizinicial[i - 1] or matrizinicial[i] == matrizinicial[i + 1] or matrizinicial[i] == matrizinicial[ancho + i]:
            matrizFinal[i] = 1
        else:
            matrizFinal[i] = 0
    if isTopRigthItem(i):
        if matrizinicial[i] == matrizinicial[i - 1] or matrizinicial[i] == matrizinicial[(ancho * 2) - 1]:
            matrizFinal[i] = 1
        else:
            matrizFinal[i] = 0
    if isLeftColumn(i):
        if matrizinicial[i] == matrizinicial[i + 1] or matrizinicial[i] == matrizinicial[i - ancho] or matrizinicial[i] == matrizinicial[i + ancho]:
            matrizFinal[i] = 1
        else:
            matrizFinal[i] = 0
    if isRigthColumn(i):
        if matrizinicial[i] == matrizinicial[i - 1] or matrizinicial[i] == matrizinicial[i - ancho] or matrizinicial[i] == matrizinicial[i + ancho]:
            matrizFinal[i] = 1
        else:
            matrizFinal[i] = 0
    if isBottomLeftItem(i):
        if matrizinicial[i] == matrizinicial[i + 1] or matrizinicial[i] == matrizinicial[len(matrizinicial) - (ancho * 2)]:
            matrizFinal[i] = 1
        else:
            matrizFinal[i] = 0
    if isBottomRow(i):
        if matrizinicial[i] == matrizinicial[i - ancho] or matrizinicial[i] == matrizinicial[i - 1] or matrizinicial[i] == matrizinicial[i + 1]:
            matrizFinal[i] = 1
        else:
            matrizFinal[i] = 0
    if isBottomRigthItem(i):
        if matrizinicial[i] == matrizinicial[i - 1] or matrizinicial[i] == matrizinicial[i - ancho]:
            matrizFinal[i] = 1
        else:
            matrizFinal[i] = 0
    if isCenterItem(i):
        if matrizinicial[i] == matrizinicial[i - 1] or matrizinicial[i] == matrizinicial[i + 1] or matrizinicial[i] == matrizinicial[i - ancho] or matrizinicial[i] == matrizinicial[i + ancho]:
            matrizFinal[i] = 1
        else:
            matrizFinal[i] = 0

matriz2 = np.reshape(matrizFinal, (alto, ancho), order="C")

for i in range(alto):
    print(matriz1[i, :],
          (matriz2[i, :]))
```
